```python
# function to repeat the exploring process to all the neighborhoods in Toronto
import urllib
def getNearbyVenues(names, latitudes, longitudes, radius=1000, categoryIds=''):
    try:
        venues_list=[]
        for name, lat, lng in zip(names, latitudes, longitudes):
            #print(name)

            # create the API request URL
            url = 'https://api.foursquare.com/v2/venues/search?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(CLIENT_ID, CLIENT_SECRET, VERSION, lat, lng, radius, LIMIT)

            if (categoryIds != ''):
                url = url + '&categoryId={}'
                url = url.format(categoryIds)

            # make the GET request
            response = requests.get(url).json()
            results = response["response"]['venues']

            # return only relevant information for each nearby venue
            for v in results:
                success = False
                try:
                    category = v['categories'][0]['name']
                    success = True
                except:
                    pass

                if success:
                    venues_list.append([(
                        name, 
                        lat, 
                        lng, 
                        v['name'], 
                        v['location']['lat'], 
                        v['location']['lng'],
                        v['categories'][0]['name']
                    )])

        nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
        nearby_venues.columns = ['Munincipality', 
                  'Munincipality Latitude', 
                  'Munincipality Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    except:
        print(url)
        print(response)
        print(results)
        print(nearby_venues)

    return(nearby_venues)
```

```python
# function to add markers for given venues to map
def addToMap(df, color, existingMap):
    for lat, lng, local, venue, venueCat in zip(df['Venue Latitude'], df['Venue Longitude'], df['Munincipality'], df['Venue'], df['Venue Category']):
        label = '{} ({}) - {}'.format(venue, venueCat, local)
        label = folium.Popup(label, parse_html=True)
        folium.CircleMarker(
            [lat, lng],
            radius=5,
            popup=label,
            color=color,
            fill=True,
            fill_color=color,
            fill_opacity=0.7).add_to(existingMap)
```

```python
def addColumn(startDf, columnTitle, dataDf):
    grouped = dataDf.groupby('Munincipality').count()
    
    for n in startDf['Munincipality']:
        try:
            startDf.loc[startDf['Munincipality'] == n,columnTitle] = grouped.loc[n, 'Venue']
        except:
            startDf.loc[startDf['Munincipality'] == n,columnTitle] = 0
```

```python
df_data = Athens_Data.copy()
df_data.rename(columns={'Munincipalities':'Munincipality'}, inplace=True)
addColumn(df_data, 'Pizza Places', Athens_venues_Pizza)
addColumn(df_data, 'High Schools', Athens_highschools)
addColumn(df_data, 'Universities', Athens_universities)
addColumn(df_data, 'Offices', Athens_office)
df_data
```

```python
# negative score, because the customer wants to open a Pizza place with the minimum competition
score_pizza = -1

# positive score, because high school students are good customers but with small pockets
score_schools = 0.5

# positive score, because uni students are good customers with medium pockets
score_uni = 1

# positive score because employees are the best customers
score_offices = 2
```

```python
df_score = df_data[['Munincipality']].copy()
```

```python
df_score['Score'] = df_data['Pizza Places'] * score_pizza + df_data['High Schools'] * score_schools + df_data['Universities'] * score_uni + df_data['Offices'] * score_offices
df_score = df_score.sort_values(by=['Score'], ascending=False)
df_score
```
